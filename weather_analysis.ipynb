# UrbanHub - Analyse Météo Historique (Phase 1)

Ce carnet permet d'explorer les données météo nettoyées au format Parquet.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os

# Configuration style
plt.style.use('ggplot')
sns.set_palette('husl')
```

## 1. Chargement des données Parquet

```python
files = glob.glob('Data/Clean/weather_*.parquet')
print(f"Fichiers trouvés : {len(files)}")

df = pd.concat([pd.read_parquet(f) for f in files])
df['timestamp'] = pd.to_datetime(df['timestamp'])
print(f"Total lignes chargées : {len(df):,}")
```

## 2. Évolution Temporelle Mensuelle (2009-2025)

```python
df['year_month'] = df['timestamp'].dt.to_period('M')
trend = df.groupby('year_month')['temperature'].mean().reset_index()
trend['year_month'] = trend['year_month'].dt.to_timestamp()

plt.figure(figsize=(15, 6))
plt.plot(trend['year_month'], trend['temperature'], color='tab:red', linewidth=1.5)
plt.title("Température moyenne mensuelle en France")
plt.xlabel("Temps")
plt.ylabel("Température (°C)")
plt.show()
```

## 3. Top 10 des stations les plus venteuses

```python
top_wind = df.groupby('station_id')['wind_speed'].max().sort_values(ascending=False).head(10)
top_wind
```
